```python
# Import required libraries/modules

import pandas as pd
import os
import warnings as wn
wn.filterwarnings('ignore')
```

```python
# Change working directory to Downloads

os.chdir(os.getcwd() + '\Downloads')
```

```python
# select the relevant columns

df = pd.read_csv('countries-table.csv') \
    [['country', 'landAreaKm', 'growthRate', 'netChange', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']]

df
```

```python
# Check the number of rows and columns (rows, columns)

df.shape
```

```python
# List out the columns in the dataset

df.columns
```

```python
# check the data types of the columns

df.dtypes
```

```python
# Display the netChange and growthRate per country

df.groupby('country', as_index = False)[('netChange', 'growthRate')].sum()
```

```python
# Display the landAreaKm per country

df.groupby('country', as_index = False)['landAreaKm'].mean()
```

```python
# Plot the landAreaKm per country as a pie chart

df.groupby('country')['landAreaKm'].mean().plot(kind = 'pie', figsize = (30,10))
```

```python
# Explore simple summary stats

df[['landAreaKm', 'growthRate']].describe(include='all')
```

```python
# look at the dataframe once again, this time, the tiny countries

df.sort_values('landAreaKm', ascending=False).tail(20)
```

```python
# We chose to filter off countries that are less than 50,000 in land area size

data = df[df['landAreaKm']  > 50000]

data
```

```python
# Explore simple summary stats once more

data[['landAreaKm', 'growthRate']].describe(include='all')
```

```python
# Examine the upper quartile of land area

data['landAreaKm'].quantile(0.75)
```

```python
# Examine the upper quartile of growth rate

data['growthRate'].quantile(0.75)
```

```python
# What are the top 5 largest countries whose growth rate are in the top 25%?

data[data['growthRate'] >= data['growthRate'].quantile(0.75)].sort_values('landAreaKm', ascending=False).head()
```

```python
# What are the top 5 most growing countries whose size are in the top 25%?

data[data['landAreaKm'] >= data['landAreaKm'].quantile(0.75)].sort_values('growthRate', ascending=False).head()
```

```python
# Examine the countries with highest growth rates and land areas

data.sort_values(['growthRate', 'landAreaKm'], ascending=False).head()
```

```python
# Considering the top 5 most growing countries whose sizes are in the top 25%

top_growth_land = data[data['landAreaKm'] >= data['landAreaKm'].quantile(0.75)] \
                .sort_values('growthRate', ascending=False).head()

top_growth_land
```

```python
# let's visualize a line plot comparison of population trends per year per country
## first we have to select the relevant columns for this plot

top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']]
```

```python
# Then we plot comparison of population trends per year per country as line plot

top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']] \
    .plot.line(x='country', figsize=(15, 8))
```

```python
# The above plot is not clear
## We need to make the country become legend, so that the population becomes values
### to do this, we need to transpose the dataframe before passing it to the plot.line() method
```

```python
# However to pass this dataframe to the plot.line() method we need to transpose it, 
## but we must promote the first column to become an index before we transpose
### this is how it appears when we set index to country:

top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']].set_index('country')
```

```python
# This is the transposed dataframe

top_growth_land_trans = top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', \
                                         'pop2023', 'pop2030', 'pop2050']] \
                                        .set_index('country').transpose()

top_growth_land_trans
```

```python
# Before we plot, we can just add a column for the average population per year for comparison sake

top_growth_land_trans['AVG_POP'] = (top_growth_land_trans['Niger'] \
                                + top_growth_land_trans['DR Congo'] \
                                + top_growth_land_trans['Chad'] \
                                + top_growth_land_trans['Mali'] \
                                + top_growth_land_trans['Angola']) / 5

top_growth_land_trans
```

```python
# Visualizing the top 5 most growing countries whose sizes are in the top 25%

top_growth_land_trans.plot(kind = 'line', figsize=(15, 8), title='population of countries by year', \
                color={'AVG_POP':'black', 'Niger':'red', 'DR Congo':'yellow', 'Chad':'green', 'Mali':'blue', 'Angola':'orange'})
```

```python
# Visualizing the Population by land Area in Km sq of the top 5 most growing countries whose size are in the top 25%

top_growth_land.plot(kind = 'barh', x='country', y='landAreaKm', title='Population by land Area in Km sq', \
                     legend=False, figsize=(15, 6), color=['green', 'orange', 'red', 'blue', 'yellow'])
```

```python
# Visualizing the Growth rate per country of the top 5 most growing countries whose size are in the top 25%

top_growth_land.plot(kind = 'barh', x='country', y='growthRate', title='Growth rate per country', \
                     legend=False, figsize=(15, 6))
```

```python
# Considering the top 5 largest countries whose growth rate are in the top 25%

top_growth_land = data[data['growthRate'] >= data['growthRate'].quantile(0.75)].sort_values('landAreaKm', ascending=False).head()

top_growth_land
```

```python
# Select the country column as well as the yearly population columns

top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']]
```

```python
# Let's set the country column as index

top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']].set_index('country')
```

```python
# Then let's transpose

top_growth_land_trans = top_growth_land[['country', 'pop1980', 'pop2000','pop2010', 'pop2022', 'pop2023', 'pop2030', 'pop2050']] \
                    .set_index('country').transpose()

top_growth_land_trans
```

```python
# We then plot the yearly population trend per countries selected

top_growth_land_trans.plot(kind = 'line', figsize = (16,8), title = 'population per country', 
                    color = ['red','black','yellow','green','orange'])
```
